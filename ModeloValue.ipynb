```python
import pandas as pd
import numpy as np
import joblib
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.ensemble import HistGradientBoostingRegressor
```

```python
data = pd.read_csv("C:/Users/CHZ/Desktop/valuecontrol.csv")
```

```python
data.head()
```

```python
columna_objetivo = 'Value'

columnas_numericas = data.select_dtypes(include='number').columns

corr_with_target = data[columnas_numericas].corr()[columna_objetivo]

sorted_columns = corr_with_target.sort_values()

print(f"Listado de columnas numéricas con menor correlación con {columna_objetivo}")
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(sorted_columns)
```

```python
X = data[['Min', 'Starts', 'MP', 'CarPrgDist', 'CarTotDist', 'Rec' ,'Carries','SCA','TouAtt3rd','Shots','PPA']] 
y = data['Value']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)

model = HistGradientBoostingRegressor()
model.fit(X, y)

y_pred = model.predict(X_test)

r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f'R^2 Score: {r2}')
print(f'Mean Squared Error: {mse}')
```
